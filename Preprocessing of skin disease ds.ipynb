```python
base_skin_dir = r"/content/ISIC2018_Task3_Training_Input/HAM10000_images"
skin_df = pd.read_csv("/content/ISIC2018_Task3_Training_Input/HAM10000_metadata.csv") # load in the data
skin_df.head()
```

```python
base_skin_dir
```

```python
imageid_path_dict = {os.path.splitext(os.path.basename(x))[0]: x
                     for x in glob(os.path.join(base_skin_dir, '*.jpg'))}
```

```python
lesion_type_dict = {
    'nv': 'Melanocytic_nevi',
    'mel': 'melanoma',
    'bkl': 'Benign_keratosis-like_lesions',
    'bcc': 'Basal_cell_carcinoma',
    'akiec': 'Actinic_keratoses',
    'vasc': 'Vascular_lesions',
    'df': 'Dermatofibroma'
}

lesion_danger = {
    'nv': 0, # 0 for benign
    'mel': 1, # 1 for malignant
    'bkl': 0, # 0 for benign
    'bcc': 1, # 1 for malignant
    'akiec': 1, # 1 for malignant
    'vasc': 0,
    'df': 0
}
```

```python
imageid_path_dict
```

```python
skin_df["path"] = skin_df["image_id"].map(imageid_path_dict.get) # map image_id to the path of that image
```

```python
skin_df["cell_type"] = skin_df["dx"].map(lesion_type_dict.get) # map dx to type of lesion
```

```python
skin_df.head()
```

```python
skin_df.head()
```

```python
skin_df["cell_type_idx"] = pd.Categorical(skin_df["cell_type"]).codes # give each cell type a category id
```

```python
skin_df.head()
```

### Reshape image and get data for classification

```python
from PIL import Image
```

### Resize image

```python
reshaped_image = skin_df["path"].map(lambda x: np.asarray(Image.open(x).resize((224,224), resample=Image.LANCZOS).\
                                                          convert("RGB")))
```

```python
out_vec = np.stack(reshaped_image, 0)
```

```python
out_vec.shape
```

```python
out_vec = out_vec.astype("float32")
out_vec /= 255
```

```python
out_vec
```

```python
labels = skin_df["cell_type_idx"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_orig, X_test, y_train_orig, y_test = train_test_split(out_vec, labels, test_size=0.25,random_state=42)
```

```python
import os
from glob import glob
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding

from keras.preprocessing.image import ImageDataGenerator
from keras import layers
from keras import Model
from keras.applications.densenet import DenseNet201
from tensorflow.keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau
import keras.backend as K
from keras.models import load_model
%matplotlib inline
import matplotlib.pyplot as plt
import gc
```

```python
X = X_train_orig
y=y_train_orig
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33333, random_state=42)
```

```python
X_train.shape, X_val.shape
```

```python
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
```

```python
class_names = np.array(['akiec', 'bcc', 'bkl', 'df', 'nv', 'vasc', 'mel'])
print(class_names)
```

```python
pre_trained_model = DenseNet201(input_shape=(224, 224, 3), include_top=False, weights="imagenet")
```

```python
for layer in pre_trained_model.layers:
    print(layer.name)
    layer.trainable = False
    
print(len(pre_trained_model.layers))
```

```python
last_layer = pre_trained_model.get_layer('relu')
print('last layer output shape:', last_layer.output_shape)
last_output = last_layer.output
```

```python
# Flatten the output layer to 1 dimension
x = layers.GlobalMaxPooling2D()(last_output)
# Add a fully connected layer with 512 hidden units and ReLU activation
x = layers.Dense(512, activation='relu')(x)
# Add a dropout rate of 0.7
x = layers.Dropout(0.5)(x)
# Add a final sigmoid layer for classification
x = layers.Dense(7, activation='softmax')(x)

# Configure and compile the model

model = Model(pre_trained_model.input, x)
optimizer = Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=True)
model.compile(loss='categorical_crossentropy',
              optimizer=optimizer,
              metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator(rotation_range=60, width_shift_range=0.2, height_shift_range=0.2,
                                   shear_range=0.2, zoom_range=0.2, fill_mode='nearest')

train_datagen.fit(X_train)

val_datagen = ImageDataGenerator()
val_datagen.fit(X_val)
```

```python
batch_size = 32
epochs = 3
history = model.fit(train_datagen.flow(X_train,y_train, batch_size=batch_size),
                              epochs = epochs, validation_data = val_datagen.flow(X_val, y_val),
                              verbose = 1, steps_per_epoch=(X_train.shape[0] // batch_size), 
                              validation_steps=(X_val.shape[0] // batch_size))
```

```python
for layer in model.layers:
    layer.trainable = True
```

```python
optimizer = Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='categorical_crossentropy',
              optimizer=optimizer,
              metrics=['acc'])
```

```python
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', patience=3, verbose=1, factor=0.5, 
                                            min_lr=0.000001, cooldown=2)
```

```python
model.summary()
```

```python
batch_size = 32
epochs = 40
history = model.fit(train_datagen.flow(X_train,y_train, batch_size=batch_size),
                              epochs = epochs, validation_data = val_datagen.flow(X_val, y_val),
                              verbose = 1, steps_per_epoch=(X_train.shape[0] // batch_size),
                              validation_steps=(X_val.shape[0] // batch_size),
                              callbacks=[learning_rate_reduction])
```

```python
loss_val, acc_val = model.evaluate(X_val, y_val, verbose=1)
print("Validation: accuracy = %f  ;  loss_v = %f" % (acc_val, loss_val))
```

```python
# Retrieve a list of accuracy results on training and test data
# sets for each training epoch
acc = history.history['acc']
val_acc = history.history['val_acc']

# Retrieve a list of list results on training and test data
# sets for each training epoch
loss = history.history['loss']
val_loss = history.history['val_loss']

# Get number of epochs
epochs = range(len(acc))

# Plot training and validation accuracy per epoch
plt.plot(epochs, acc, label = "training")
plt.plot(epochs, val_acc, label = "validation")
plt.legend(loc="upper left")
plt.title('Training and validation accuracy')

plt.figure()

# Plot training and validation loss per epoch
plt.plot(epochs, loss, label = "training")
plt.plot(epochs, val_loss, label = "validation")
plt.legend(loc="upper right")
plt.title('Training and validation loss')
```
